````python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the above dataframe is stored in a variable named 'df'
df.sort_values(['name', 'year'], inplace=True)

# Calculate the year-over-year income growth rate
df['income_growth'] = df.groupby('name')['income'].pct_change()

# Filter the dataframe for the year 2019 only
df_2019 = df[df['year'] == 2019]

# Plot the income growth for each state in a bar chart
plt.figure(figsize=(10,6))
plt.bar(df_2019['name'], df_2019['income_growth'])
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Income Growth Rate')
plt.title('Year-over-year Income Growth Rate by State (2019)')
plt.show()
```
Please note that the `pct_change()` function calculates the percentage change between the current and a prior element. This means the income growth rate is calculated based on the previous year's income for each state. The result is then multiplied by 100 to get the percentage. The bar chart is plotted for the year 2019 only.
````
